## Monte Carlo Method

```python
# HIDDEN
import matplotlib.pyplot as plt
import random
import numpy as np
import time

def estimate_pi(num_points):
    inside_circle = 0
    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1
    return 4 * inside_circle / num_points

def plot_points(points_in_circle, points_outside_circle, estimated_pi, num_points, index):
    theta = np.linspace(0, np.pi/2, 100)
    x_boundary = np.cos(theta)
    y_boundary = np.sin(theta)

    plt.subplot(1, 3, index + 1)
    plt.scatter(*zip(*points_in_circle), color='blue', s=1, label='Inside Circle')
    plt.scatter(*zip(*points_outside_circle), color='red', s=1, label='Outside Circle')
    plt.plot(x_boundary, y_boundary, color='green', label='Circle Boundary')
    plt.xlim(0, 1)
    plt.ylim(0, 1)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.title(f'Estimation of Pi (num_points={num_points:.0e})')
    plt.xlabel('X')
    plt.ylabel('Y')

    # Print estimated value and error with white background
    plt.text(0.05, 0.75, f'Estimated Pi: {estimated_pi:.6f}', fontsize=12, color='black',
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    plt.text(0.05, 0.65, f'Error: {abs(estimated_pi - np.pi):.6f}', fontsize=12, color='black',
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    plt.text(0.05, 0.55, f'Runtime: {runtimes[index]:.6f} seconds', fontsize=12, color='black',
             bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))

# Define the number of points for each plot
num_points_list = [100, 100000, 100000000]
runtimes = []

# Generate random points, estimate pi, and plot for each num_points
plt.figure(figsize=(15, 5))

for index, num_points in enumerate(num_points_list):
    points_in_circle = []
    points_outside_circle = []

    start_time = time.time()  # Start measuring time

    for _ in range(num_points):
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        if x**2 + y**2 <= 1:
            points_in_circle.append((x, y))
        else:
            points_outside_circle.append((x, y))

    estimated_pi = estimate_pi(num_points)

    end_time = time.time()  # End measuring time
    runtime = end_time - start_time
    runtimes.append(runtime)

    # Print the estimated value of pi and error
    print(f"Estimated value of pi (num_points={num_points:.0e}): {estimated_pi}")
    print(f"Error (num_points={num_points:.0e}): {abs(estimated_pi - np.pi)}")
    print(f"Runtime (num_points={num_points:.0e}): {runtime} seconds")

    # Create subplots
    plot_points(points_in_circle, points_outside_circle, estimated_pi, num_points, index)

# Adjust layout
plt.tight_layout()

# Show all plots
plt.show()
```
